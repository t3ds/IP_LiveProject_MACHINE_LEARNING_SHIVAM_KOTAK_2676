```python
import pandas as pd
import itertools
```

```python
dataset = pd.read_excel("CCMLEmployeeData_2.xlsx")
dataset
```

```python
dataset.head(24)
```

```python
domains = dataset.DomainU.dropna()
```

```python
domains = domains.to_list()
domains
```

```python
events = dataset.EventU.dropna()
```

```python
events = events.to_list()
```

```python
events
```

```python
prod = itertools.product(domains, events)
print(type(prod))
```

```python
pool = [domains, events]
temp = []
for i in itertools.product(*pool):
    temp.append('.'.join(i))
    
print(temp)
```

```python
with open('labels.txt', 'w') as f: 
    for i in temp:
        f.write(i + '\n')
```
